```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def etiquetar(image):
    _,image_bin = cv2.threshold(image,127,255,cv2.THRESH_BINARY)

    contorno,_ = cv2.findContours(image_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    img_et = np.zeros_like(image, dtype=np.uint8)

    for i, contorno1 in enumerate(contorno):
        cv2.drawContours(img_et, [contorno1], -1, (i+1), -1)
    return img_et

img_original = cv2.imread('nami.jpg', cv2.IMREAD_GRAYSCALE)
imagen_etiquetada = etiquetar(img_original)


plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.title('Original')
plt.imshow(img_original ,cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.title('Etiquetada')
plt.imshow(imagen_etiquetada ,cmap='jet')
plt.axis('off')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.measure as ms

img_original = cv2.imread('nami.jpg', cv2.IMREAD_GRAYSCALE)
_,image_bin = cv2.threshold(img_original,127,255,cv2.THRESH_BINARY)

img_final = ms.label(image_bin, connectivity=1)

plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.title('Original')
plt.imshow(img_original ,cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.title('Etiquetada')
plt.imshow(img_final ,cmap='jet')
plt.axis('off')
```
