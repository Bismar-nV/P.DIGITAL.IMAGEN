```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def conv_horizontal(img, mask_size):
    mask = np.ones((1, mask_size), np.float32) / mask_size
    img_conv = cv2.filter2D(img, -1, mask)
    return img_conv

image = cv2.imread('img2.jpg')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_r = conv_horizontal(img_rgb[:,:,0], 11)
img_g = conv_horizontal(img_rgb[:,:,1], 11)
img_b = conv_horizontal(img_rgb[:,:,2], 11)

img_fin = np.stack([img_r, img_g, img_b], axis=2)

plt.subplot(121)
plt.title("img original")
plt.imshow(img_rgb)

plt.subplot(122)
plt.title("img final (horizontal)")
plt.imshow(img_fin)

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def conv_vertical(img, mask_size):
    mask = np.ones((mask_size, 1), np.float32) / mask_size
    img_conv = cv2.filter2D(img, -1, mask)
    return img_conv

def conv_horizontal(img, mask_size):
    mask = np.ones((1, mask_size), np.float32) / mask_size
    img_conv = cv2.filter2D(img, -1, mask)
    return img_conv

def conv_diagonal(img, mask_size):
    mask = np.eye(mask_size, dtype=np.float32) / mask_size
    img_conv = cv2.filter2D(img, -1, mask)
    return img_conv

image = cv2.imread('img2.jpg')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

img_r = conv_vertical(img_rgb[:,:,0], 25)
img_g = conv_vertical(img_rgb[:,:,1], 25)
img_b = conv_vertical(img_rgb[:,:,2], 25)

img_r = conv_horizontal(img_rgb[:,:,0], 25)
img_g = conv_horizontal(img_rgb[:,:,1], 25)
img_b = conv_horizontal(img_rgb[:,:,2], 25)

img_r = conv_diagonal(img_rgb[:,:,0], 11)
img_g = conv_diagonal(img_rgb[:,:,1], 25)
img_b = conv_diagonal(img_rgb[:,:,2], 50)

img_fin_vertical = np.stack([img_r, img_g, img_b], axis=2)
img_fin_horizontal = np.stack([img_r, img_g, img_b], axis=2)
img_fin_diagonal = np.stack([img_r, img_g, img_b], axis=2)

plt.figure(figsize=(12, 6))

plt.subplot(221)
plt.title("img original")
plt.imshow(img_rgb)

plt.subplot(222)
plt.title("img final (vertical)")
plt.imshow(img_fin_vertical)

plt.subplot(223)
plt.title("img final (horizontal)")
plt.imshow(img_fin_horizontal)

plt.subplot(224)
plt.title("img final (diagonal)")
plt.imshow(img_fin_diagonal)

plt.tight_layout()
plt.show()
```
