1.- Atenúa el ruido en la imagen ruido.jpg tal cual se ve en el ejemplo:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def conv_horizontal(img, mask_size):
    mask = np.ones((1, mask_size), np.float32) / mask_size
    img_conv = cv2.filter2D(img, -1, mask)
    return img_conv

image = cv2.imread('ruido.jpg')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_r = conv_horizontal(img_rgb[:,:,0], 9)
img_g = conv_horizontal(img_rgb[:,:,1], 9)
img_b = conv_horizontal(img_rgb[:,:,2], 9)

img_fin = np.stack([img_r, img_g, img_b], axis=2)

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.title("img original")
plt.imshow(img_rgb)

plt.subplot(122)
plt.title("img final")
plt.imshow(img_fin)

plt.show()
```

2.- Usando la imagen “rosa.png” obtenga el siguiente resultado:


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('rosa.png')
imagen = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, mascara = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
gaus_image = cv2.GaussianBlur(imagen, (21, 21), 0)
desenfocado_image = np.where(mascara[:, :, None] == 255, imagen, gaus_image)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(imagen)
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(desenfocado_image)
plt.title('Imagen con fondo desenfocado')
plt.axis('off')

plt.show()
```

3.- Usando la imagen “Elementos jpg” obtenga el siguiente resultado.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage.measure as ms

def count_objects(img, threshold_value):
    _, binary = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY_INV)
    target = ms.label(binary, connectivity=2)
    cont = np.max(target)
    return cont, binary
imagen1 = cv2.imread('elementos.jpg', cv2.IMREAD_GRAYSCALE)
imagen_invertida = cv2.bitwise_not(imagen1)
plt.figure(figsize=(15, 6))

plt.subplot(2, 3, 1)
plt.imshow(imagen1, cmap='gray')
plt.title("Imagen Invertida")

plt.subplot(2, 3, 2)
plt.imshow(imagen_invertida, cmap='gray')
plt.title("Imagen Invertida")

plt.show()
```
