```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage.measure as ms

imagen1 = cv2.imread('elementos.jpg')
imagen_gris = cv2.cvtColor(imagen1, cv2.COLOR_BGR2GRAY)
_, binarizada = cv2.threshold(imagen_gris, 128, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
contornos, _ = cv2.findContours(binarizada, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
numero_contornos = len(contornos)
plt.figure(figsize=(20, 10))

plt.subplot(2, 4, 1)
plt.imshow(cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB))
plt.title("Imagen Original")
plt.axis('off')

plt.subplot(2, 4, 2)
plt.imshow(imagen_gris, cmap='gray')
plt.title("Imagen en Escala de Grises")
plt.axis('off')

plt.subplot(2, 4, 3)
plt.imshow(binarizada, cmap='gray')
plt.title("Imagen Binarizada")
plt.axis('off')

imagen_contornos = np.zeros_like(imagen_gris)
cv2.drawContours(imagen_contornos, contornos, -1, (255, 255, 255), 1)

plt.subplot(2, 4, 4)
plt.imshow(imagen_contornos, cmap='gray')
plt.title(f"Contornos Encontrados: {numero_contornos}")
plt.axis('off')

plt.tight_layout()
plt.show()
```
