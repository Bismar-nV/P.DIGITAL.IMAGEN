```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('nami.jpg')
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, mascara=cv2.threshold(gris,127,255,cv2.THRESH_BINARY)

kernel= np.array([[0,1,0],[1,1,1],[0,1,1]],dtype=np.uint8)#personalizable

erosion=cv2.erode(mascara, kernel, iterations=1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Imagen Original')
plt.imshow(img)

plt.subplot(2, 2, 2)
plt.title('mascara')
plt.imshow(mascara)

plt.subplot(2, 2, 3)
plt.title('erosion')
plt.imshow(erosion)
plt.show()
```

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt


img=cv2.imread('nami.jpg')
img_2=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
arr1=np.array([10,10,10])
arr2=np.array([148,127,100])

copia=img_2.copy()

mascara=cv2.inRange(copia,arr1,arr2)
kernel=np.ones((3,3),np.uint8)
erosion1=cv2.erode(mascara,kernel , iterations=1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Imagen Original')
plt.imshow(img)

plt.subplot(2, 2, 2)
plt.title('mascara')
plt.imshow(mascara)

plt.subplot(2, 2, 3)
plt.title('erosion')
plt.imshow(erosion1)
plt.show()
```
