```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('nami.jpg', 0)

list = [
    np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]),
    np.array([[0, -1, 0], [0, 5, 0], [0, -1, 0]]),
    np.array([[0, 0, 0], [-1, 5, -1], [0, 0, 0]]),
    np.array([[-1, 0, -1], [0, 5, 0], [-1, 0, -1]]),
    np.array([[-1, -1, -1], [0, 5, 0], [0, 0, 0]]),
    np.array([[0, 0, -1], [0, 5, 0], [-1, 0, 0]]),
    np.array([[0, -1, 0], [0, 5, -1], [0, 0, 0]]),
    np.array([[0, -1, 0], [0, 5, 0], [0, -1, 0]]),
    np.array([[0, 0, 0], [-1, 5, 0], [0, 0, -1]]),
    np.array([[-1, 0, 0], [0, 5, 0], [0, 0, -1]]),
    np.array([[0, 0, 0], [-1, 5, 0], [0, -1, 0]]),
    np.array([[0, 0, -1], [0, 5, 0], [1, 0, -1]]),
    np.array([[0, -1, -1], [0, 5, 0], [0, -1, 0]]),
    np.array([[0, -1, 0], [-1, 5, -1], [0, 1, 0]]),
    np.array([[-1, 0, 0], [0, 5, -1], [0, -1, 0]]),
    np.array([[0, 0, 0], [0, 5, -1], [0, -1, 0]]),
    np.array([[0, -1, 0], [0, 5, 0], [-1, 0, 0]]),
    np.array([[0, 0, 0], [-1, 5, -1], [0, -1, 0]]),
    np.array([[-1, 0, 0], [0, 5, 0], [-1, 0, 0]]),
    np.array([[0, -1, 0], [1, 5, -1], [0, -1, 0]])
]

plt.figure(figsize=(14, 14))
for i, lista in enumerate(list):
    img_fin = cv2.filter2D(img, -1, lista)
    plt.subplot(5, 4, i + 1)
    plt.imshow(img_fin, cmap='gray')
    plt.title(f'img {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
