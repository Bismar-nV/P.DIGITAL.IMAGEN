```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
image = io.imread('img1.jpg', as_gray=True)
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
downsample_factor = 4
downsampled_image = image[::downsample_factor, ::downsample_factor]
plt.subplot(1, 2, 2)
plt.imshow(downsampled_image, cmap='gray')
plt.title('Imagen Muestreada (Aliasing)')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread('img1.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_blurred = cv2.GaussianBlur(img_rgb, (5, 55), 0)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')

plt.subplot(1, 2, 2)
plt.imshow(img_blurred)
plt.title('Imagen suavizada')

plt.show()
```
