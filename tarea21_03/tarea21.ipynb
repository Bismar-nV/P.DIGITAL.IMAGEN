```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

def aumentar_brillo(imagen, factor):
    imagen_hsv = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)
    imagen_hsv[:,:,2] = cv2.add(imagen_hsv[:,:,2], factor)
    imagen_rgb = cv2.cvtColor(imagen_hsv, cv2.COLOR_HSV2RGB)
    return imagen_rgb

def actualizar_brillo(change):
    factor_brillo = change.new
    imagen_brillante = aumentar_brillo(imagen_original.copy(), factor_brillo)
    plt.figure(figsize=(10,5))
    plt.imshow(imagen_brillante)
    plt.title('Imagen con Brillo Aumentado')
    plt.show()

# Cargar la imagen
imagen_original = cv2.imread('nami.jpg')

# Aumentar el brillo con un factor de 50
imagen_brillante = aumentar_brillo(imagen_original, 150)

# Crear widget de deslizador para ajustar el brillo
slider_brillo = widgets.IntSlider(value=0, min=0, max=100, step=1, description='Brillo:')
slider_brillo.observe(actualizar_brillo, 'value')

# Mostrar la imagen original y el widget de deslizador
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(imagen_brillante)
plt.title('Imagen con Brillo Aumentado')

plt.show()
display(slider_brillo)
```
